# Interpolate regression labels on sequential data

```
import numpy as np
```

```
from scipy import interpolate
```

```
import matplotlib.pyplot as plt
```

# Standard interpolation

https://docs.scipy.org/doc/scipy/reference/generated/scipy.interpolate.interp1d.html#scipy.interpolate.interp1d

```
#@title Interpolate regression labels on sequential data { run: "auto" }
m = 35 #@param {type:"slider", min:1, max:100, step:1}
np.random.seed(42)
X = np.random.randint(10, size=10)

f = interpolate.interp1d(np.arange(len(X)), X) 

x_i  = np.linspace(0, len(X)-1, num=m)
y_i = f(x_i)

plt.plot(range(len(X)), X)
plt.scatter(range(len(X)), X, c='g', label='True')
plt.scatter(x_i, y_i, c='r', label='Interpolated')
plt.legend()
plt.show()
```

# Splined interpolation

https://docs.scipy.org/doc/scipy/reference/generated/scipy.interpolate.pchip_interpolate.html#scipy.interpolate.pchip_interpolate

```
#@title Interpolate regression labels on sequential data { run: "auto" }
m = 40 #@param {type:"slider", min:1, max:100, step:1}
np.random.seed(42)
X = np.random.randint(10, size=10)

x_i  = np.linspace(0, len(X)-1, num=m)
y_i = interpolate.pchip_interpolate(np.arange(len(X)), X, x_i)

plt.plot(range(len(X)), X)
plt.scatter(range(len(X)), X, c='g', label='True')
plt.scatter(x_i, y_i, c='r', label='Interpolated')
plt.legend()
plt.show()
```
